### Read the Given data

```python
import psycopg2
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Connect to PostgreSQL database
conn = psycopg2.connect(
    dbname="telecom",  # Database name
    user="postgres",  # Username
    password="12345",  # Your PostgreSQL password
    host="localhost"  # Database server
)

# Query to fetch all xDR records
query = "SELECT * FROM xdr_data;"  # Adjust the table name as needed

# Load the data into a pandas DataFrame
df = pd.read_sql(query, conn)

# Close the database connection
conn.close()

# Display first few rows of the dataset to understand its structure
print(df.head())
```

```python
import sys
sys.path.append('../scripts')
from preprocess import fill_missing_values, aggregate_data, rename_columns, get_top_n_values, get_bottom_n_values, get_most_frequent_value

# Fill missing values
columns_to_fill = [
    'TCP DL Retrans. Vol (Bytes)', 'TCP UL Retrans. Vol (Bytes)',
    'Avg RTT DL (ms)', 'Avg RTT UL (ms)',
    'Avg Bearer TP DL (kbps)', 'Avg Bearer TP UL (kbps)'
]

for column in columns_to_fill:
    fill_missing_values(df, column)

# Aggregate per customer
agg_columns = {
    'TCP DL Retrans. Vol (Bytes)': 'mean',
    'TCP UL Retrans. Vol (Bytes)': 'mean',
    'Avg RTT DL (ms)': 'mean',
    'Avg RTT UL (ms)': 'mean',
    'Avg Bearer TP DL (kbps)': 'mean',
    'Avg Bearer TP UL (kbps)': 'mean',
    'Handset Type': 'first'
}

agg_data = aggregate_data(df, 'MSISDN/Number', agg_columns)

# Rename columns for clarity
new_column_names = [
    'MSISDN/Number', 'Avg TCP DL Retransmission', 'Avg TCP UL Retransmission',
    'Avg RTT DL (ms)', 'Avg RTT UL (ms)', 'Avg Throughput DL (kbps)',
    'Avg Throughput UL (kbps)', 'Handset Type'
]

rename_columns(agg_data, new_column_names)

# Print the aggregated data
print(agg_data.head())
```

### Handle missing values and aggregate per customer

```python
# Task 3.1 - 
# Fill missing values with the mean for TCP Retransmission, RTT, and Throughput
df['TCP DL Retrans. Vol (Bytes)'].fillna(df['TCP DL Retrans. Vol (Bytes)'].mean(), inplace=True)
df['TCP UL Retrans. Vol (Bytes)'].fillna(df['TCP UL Retrans. Vol (Bytes)'].mean(), inplace=True)
df['Avg RTT DL (ms)'].fillna(df['Avg RTT DL (ms)'].mean(), inplace=True)
df['Avg RTT UL (ms)'].fillna(df['Avg RTT UL (ms)'].mean(), inplace=True)
df['Avg Bearer TP DL (kbps)'].fillna(df['Avg Bearer TP DL (kbps)'].mean(), inplace=True)
df['Avg Bearer TP UL (kbps)'].fillna(df['Avg Bearer TP UL (kbps)'].mean(), inplace=True)

# Aggregate per customer
agg_data = df.groupby('MSISDN/Number').agg({
    'TCP DL Retrans. Vol (Bytes)': 'mean',
    'TCP UL Retrans. Vol (Bytes)': 'mean',
    'Avg RTT DL (ms)': 'mean',
    'Avg RTT UL (ms)': 'mean',
    'Avg Bearer TP DL (kbps)': 'mean',
    'Avg Bearer TP UL (kbps)': 'mean',
    'Handset Type': 'first'
}).reset_index()

# Rename columns for clarity
agg_data.columns = ['MSISDN/Number', 'Avg TCP DL Retransmission', 'Avg TCP UL Retransmission', 
                    'Avg RTT DL (ms)', 'Avg RTT UL (ms)', 'Avg Throughput DL (kbps)', 
                    'Avg Throughput UL (kbps)', 'Handset Type']

print(agg_data.head())
```

```python
# Top 10, Bottom 10, and Most Frequent TCP DL Retransmission
top_tcp_dl = get_top_n_values(agg_data, 'Avg TCP DL Retransmission', 10)
bottom_tcp_dl = get_bottom_n_values(agg_data, 'Avg TCP DL Retransmission', 10)
most_frequent_tcp_dl = get_most_frequent_value(agg_data, 'Avg TCP DL Retransmission')

# Top 10, Bottom 10, and Most Frequent TCP UL Retransmission
top_tcp_ul = get_top_n_values(agg_data, 'Avg TCP UL Retransmission', 10)
bottom_tcp_ul = get_bottom_n_values(agg_data, 'Avg TCP UL Retransmission', 10)
most_frequent_tcp_ul = get_most_frequent_value(agg_data, 'Avg TCP UL Retransmission')

# Print results
print("Top 10 TCP DL Retransmissions:\n", top_tcp_dl)
print("Bottom 10 TCP DL Retransmissions:\n", bottom_tcp_dl)
print("Most Frequent TCP DL Retransmissions:\n", most_frequent_tcp_dl)

print("Top 10 TCP UL Retransmissions:\n", top_tcp_ul)
print("Bottom 10 TCP UL Retransmissions:\n", bottom_tcp_ul)
print("Most Frequent TCP UL Retransmissions:\n", most_frequent_tcp_ul)
```
